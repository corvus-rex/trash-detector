```python
def calculate_iou(box1, box2):
    # YOLO format: [x, y, w, h]
    x1_min, y1_min = box1[0] - box1[2] / 2, box1[1] - box1[3] / 2
    x1_max, y1_max = box1[0] + box1[2] / 2, box1[1] + box1[3] / 2
    
    x2_min, y2_min = box2[0] - box2[2] / 2, box2[1] - box2[3] / 2
    x2_max, y2_max = box2[0] + box2[2] / 2, box2[1] + box2[3] / 2
    
    # Intersection
    inter_min_x = max(x1_min, x2_min)
    inter_min_y = max(y1_min, y2_min)
    inter_max_x = min(x1_max, x2_max)
    inter_max_y = min(y1_max, y2_max)
    
    inter_w = max(0, inter_max_x - inter_min_x)
    inter_h = max(0, inter_max_y - inter_min_y)
    inter_area = inter_w * inter_h
    
    # Union
    area1 = box1[2] * box1[3]
    area2 = box2[2] * box2[3]
    union_area = area1 + area2 - inter_area
    
    # IoU
    iou = inter_area / union_area if union_area > 0 else 0
    return iou

# Example
box1 = [0.5, 0.5, 0.2, 0.2]  # x, y, w, h
box2 = [0.6, 0.6, 0.2, 0.2]
print("IoU:", calculate_iou(box1, box2))
```

```python
def get_global(box1, boxl):
    x1, y1, w1, h1 = box1
    xl, yl, wl, hl = boxl
    x = (x1 - 0.5) * wl + xl
    y = (y1 - 0.5) * hl + yl
    w = w1 * wl
    h = h1 * hl
    return x, y, w, h
```

```python
trash = []

for i in range(len(df)):
    data = df.iloc[i]
    
    px, py, pw, ph = [data[f'person_{i}'] for i in ['x','y','w','h']]
    
    hand_x = [data[f'landmark_x_{i}'] for i in [15,17,19,21]]
    hand_y = [data[f'landmark_y_{i}'] for i in [15,17,19,21]]
    
    if np.isnan(hand_x[0]):
        tx = 0.5 - np.random.random() * 0.25
        ty = 0.5 + np.random.random() * 0.25
        tw = np.random.random() * 0.25
        th = np.random.random() * 0.25 * pw / ph
        x, y, w, h = get_global((tx,ty,tw,th), (px,py,pw,ph))
        trash.append([x,y,w,h])
    else:
        hand_x1, hand_x2 = min(hand_x), max(hand_x)
        hand_y1, hand_y2 = min(hand_y), max(hand_y)
        tx = (hand_x1 + hand_x2) / 2
        ty = (hand_y1 + hand_y2) / 2
        tw = np.abs(hand_x1 - hand_x2)
        th = np.abs(hand_y1 - hand_y2)
        x, y, w, h = get_global((tx,ty,tw,th), (px,py,pw,ph))
        trash.append([x,y,w,h])

cols = [f'trash_{i}' for i in ['x','y','w','h']]
df[cols] = np.array(trash)
df.head()
```

```python
stat = []

for i in range(len(df)):
    data = df.iloc[i]
    
    px, py, pw, ph = [data[f'person_{i}'] for i in ['x','y','w','h']]
    ntx, nty, ntw, nth = [data[f'nearest_trash_{i}'] for i in ['x','y','w','h']]
    tx, ty, tw, th = [data[f'trash_{i}'] for i in ['x','y','w','h']]
    
    if np.isnan(ntx):
        ratio_area = (tw*th)/(pw*ph)
        cdist = np.sqrt((tx-px)**2 + (ty-py)**2)
        cdist_rel = cdist / ph
        stat.append([ratio_area, cdist_rel])
    else:
        ratio_area = (ntw*nth)/(pw*ph)
        cdist = np.sqrt((ntx-px)**2 + (nty-py)**2)
        cdist_rel = cdist / ph
        stat.append([ratio_area, cdist_rel])

cols = ['trash_ratio_area', 'trash_rel_dist']
df[cols] = np.array(stat)
df.head()
```

```python
path = 'seq_track.pkl'
with open(path, 'rb') as f:
    seq_track = pickle.load(f)
```

```python
df_mapper = {}

feat_cols = [f'landmark_x_{i}' for i in range(33)] + [f'landmark_y_{i}' for i in range(33)] + ['trash_ratio_area', 'trash_rel_dist']

for i in range(len(df)):
    data = df.iloc[i]
    seq = data['sequence_id']
    fr = data['frame_id']
    pr = data['unproc_person_id']
    feat = data[feat_cols].values
    label = data['label']
    
    if seq not in df_mapper:
        df_mapper[seq] = {}
    if fr not in df_mapper[seq]:
        df_mapper[seq][fr] = {}
    if pr not in df_mapper[seq][fr]:
        df_mapper[seq][fr][pr] = {
        'feat': feat,
        'label': label
    }
```

```python
X = []
y = []
for seq, flabel in tqdm(enumerate(seq_track), total=len(seq_track)):
    for track in flabel.values():
        xs = []
        ys = []
        for frame_id, person_id in track:
            data = df_mapper[seq][frame_id][person_id]
            feat = data['feat']
            label = data['label']
            xs.append(feat)
            ys.append(label)
        X.append(xs)
        cnt = np.sum(ys)
        real_label = 1 if (cnt / len(ys)) > 0.5 else 0
        y.append(real_label)
```

```python
X_interpolated = []

for x in X:
    xdf = pd.DataFrame(x)
    xdf = xdf.interpolate(method='linear', axis=0, limit_direction='both')
    X_interpolated.append(xdf.values)
```

```python
max_len = max([len(x) for x in X_interpolated])
max_len
```

```python
from scipy.interpolate import interp1d

X_stretched = []

for x in tqdm(X_interpolated, total=len(X_interpolated)):
    n, k = x.shape
    original_indices = np.linspace(0, n - 1, n)
    target_indices = np.linspace(0, n - 1, max_len)
    interpolated_array = np.zeros((max_len, k))
    for col in range(k):
        interp_func = interp1d(original_indices, x[:, col], kind='linear')
        interpolated_array[:, col] = interp_func(target_indices)
    X_stretched.append(interpolated_array)
```

```python
idx = []
for i, x in enumerate(X_stretched):
    if np.isnan(x).any():
        idx.append(i)
        
len(idx)
```

```python
X_used = []
y_used = []

for i, (x, _y) in enumerate(zip(X_stretched, y)):
    if i not in idx:
        X_used.append(x)
        y_used.append(_y)

len(y_used)
```

## Training

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# Define LSTM classifier model
class LSTMClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=2, dropout_rate=0.5):
        super(LSTMClassifier, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.conv1 = nn.Conv1d(in_channels=input_size, out_channels=64, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.pool = nn.MaxPool1d(kernel_size=2)
        self.lstm = nn.LSTM(128, hidden_size, num_layers, batch_first=True)
        self.fc1 = nn.Linear(hidden_size, 128)
        self.fc2 = nn.Linear(128, output_size)
        self.dropout = nn.Dropout(dropout_rate)

    def forward(self, x):
        x = x.permute(0, 2, 1)
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        
        x = x.permute(0, 2, 1)
        h0 = torch.zeros(self.lstm.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.lstm.num_layers, x.size(0), self.hidden_size).to(x.device)
        out, _ = self.lstm(x, (h0, c0))
        out = F.relu(self.fc1(out[:, -1, :]))
        out = self.dropout(out)
        out = self.fc2(out)
        return out
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    np.array(X_used), np.array(y_used), 
    random_state=3407, stratify=y_used, test_size=0.2
)
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
```

```python
input_size = X_train.shape[2]
hidden_size = 128
num_layers = 3
output_size = 2
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = LSTMClassifier(input_size, hidden_size, output_size, num_layers).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

X_test_tensor = torch.tensor(X_test, dtype=torch.float32)

def evaluate(model):
    model.eval()
    with torch.no_grad():
        outputs = model(X_test_tensor)
        predictions = torch.argmax(outputs, dim=1).numpy()
        acc = accuracy_score(y_test, predictions)
        pre = precision_score(y_test, predictions, average='weighted', labels=np.unique(predictions))
        rec = recall_score(y_test, predictions, average='weighted', labels=np.unique(predictions))
        f1 = f1_score(y_test, predictions, average='weighted', labels=np.unique(predictions))
        
    model.train()
    return acc, pre, rec, f1
```

```python
num_epochs = 50
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()
    acc, pre, rec, f1 = evaluate(model)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}, Acc: {acc:.4f}, Pre: {pre:.4f}, Rec: {rec:.4f}, F1: {f1:.4f}')
```
